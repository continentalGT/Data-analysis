```python
fig=plt.subplots(2,2,figsize=(10,10))
plt.subplot(221)
sns.scatterplot(x=penguins1['body_mass_g'],y=penguins1['flipper_length_mm'],hue=penguins1['species'])
plt.title('mass vs flipper')


plt.subplot(222)
sns.scatterplot(x=penguins1['body_mass_g'],y=penguins1['bill_depth_mm'],hue=penguins1['species'])
plt.title('mass vs bill_depth_mm')

plt.subplot(223)
sns.scatterplot(x=penguins1['body_mass_g'],y=penguins1['bill_length_mm'],hue=penguins1['species'])
plt.title('mass vs bill_length_mm')

plt.subplot(224)
sns.scatterplot(x=penguins1['bill_depth_mm'],y=penguins1['bill_length_mm'],hue=penguins1['species'])
plt.title('bill_depth_mm vs bill_length_mm')

```

```python
fig=plt.subplots(2,2,figsize=(10,10))
plt.subplot(221)
sns.scatterplot(x=penguins1['body_mass_g'],y=penguins1['flipper_length_mm'],hue=penguins1['sex'])
plt.title('mass vs flipper')


plt.subplot(222)
sns.scatterplot(x=penguins1['body_mass_g'],y=penguins1['bill_depth_mm'],hue=penguins1['sex'])
plt.title('mass vs bill_depth_mm')

plt.subplot(223)
sns.scatterplot(x=penguins1['body_mass_g'],y=penguins1['bill_length_mm'],hue=penguins1['sex'])
plt.title('mass vs bill_length_mm')

plt.subplot(224)
sns.scatterplot(x=penguins1['bill_depth_mm'],y=penguins1['bill_length_mm'],hue=penguins1['sex'])
plt.title('bill_depth_mm vs bill_length_mm')

```
