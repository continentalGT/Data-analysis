```python
from wordcloud import WordCloud,STOPWORDS
```

```python
import wikipedia
from PIL import Image
```

```python
#for any modeule not mentioned used-'>pip install module name'
example-pip install wikipedia
```

```python
stop_w=set(STOPWORDS)
info=wikipedia.summary('Python (programming language)')
print(info)
```

```python
word_cloud=WordCloud(width = 800, height = 800,background_color='white',stopwords=stop_w).generate(info)
```

```python
#through img
img=word_cloud.to_image()
img.show()
```

```python
#trough matplotlib
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,8),facecolor = None)
plt.imshow(word_cloud)
plt.axis("off")
plt.show()
```
